## La bibliothèque Matplotlib

Les représentations graphiques jouent un rôle important dans l'analyse de données. Le language Python s'appuie sur une bibliothèque Matplotlib, qui permet de construire une grande variété de représentations graphiques, dont un aperçu est donné dans la page catalogue [dédiée](https://matplotlib.org/stable/gallery/index.html).

Nous allons explorer quelques représentations parmis les plus courantes.

```python
# On importe les bibliothèques Pandas et Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lecture du tableau des données
hd = pd.read_csv("heart.csv")
```

### 1. Histogrammes

On souhaite vérifier si les ages des patients échantillonnent correctement la population. On veut obtenir la distribution des valeurs des ages. Cette opération est réalisée par la méthode hist().

```python
# On définit un cadre pour tracer la figure
# ax est un objet graphique qui permet de personaliser les paramètres du graphique

ax = hd.age.hist(bins=40)
ax.set_xlabel('ages')
ax.set_ylabel('Frequences')
```

La commande dir(ax) permet d'obtenir une liste de toutes les méthodes qui sont présentes dans un objet donné (ici les axes d'un graphique). Prenez le temps d'explorer les possibilités offertes par les méthode dont le nom débute par set...

Note: un certains nombre de paramètres peuvent être donnés dans l'appel de la fonction hist comme la couleur (paramètre color) ou la transparence du tracé (alpha).

### 2. Mise en évidence de données corrélées

On peut mettre en évidence une éventuelle corrélation entre 2 paramètres en tracant une valeur en fonction de l'autre. Ce type de diagramme est appelé scatter plot.

```python
# Diagramme de corrélation entre l'age des patients et le taux de cholestérol 
ax = hd.plot.scatter(x='age',y='chol')
```

On peut visualiser plus de deux variables sur un graphique de corrélation en représentant un paramètre sous une forme graphique, par exemple la taille des points, ou la couleur :


```python
ax = hd.plot.scatter(x='age',y='chol',s=hd['thal']*20)
```

```python
ax = hd.plot.scatter(x='age',y='chol',c='thal',cmap='bwr')
```
